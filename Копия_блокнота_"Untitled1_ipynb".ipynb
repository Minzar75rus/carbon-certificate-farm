```python
import random
import json
import os
from datetime import datetime
import matplotlib.pyplot as plt
from google.colab import files

def get_next_cert_number():
    """Получить следующий номер сертификата"""
    counter_file = "cert_counter.txt"

    if os.path.exists(counter_file):
        with open(counter_file, 'r') as f:
            last_num = int(f.read().strip())
    else:
        last_num = 0

    next_num = last_num + 1

    # Сохраняем новый номер
    with open(counter_file, 'w') as f:
        f.write(str(next_num))

    return f"{next_num:08d}"  # Формат: 00000001

def carbon_certificate():
    # Цвета
    GREEN = '\033[92m'
    BLUE = '\033[94m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    RESET = '\033[0m'

    print(f"{GREEN}{BOLD}🌿 CARBON CERTIFICATE FARM 🌿{RESET}\n")

    plant = input(f"{BLUE}Растение (hemp / bamboo / tree): {RESET}").strip().lower()
    state = input(f"{BLUE}Биомасса: свежая (fresh) или сухая (dry)? {RESET}").strip().lower()
    kg = float(input(f"{BLUE}Вес (кг): {RESET}"))

    # Пересчёт
    dry_coeff = 0.3
    if state == "fresh":
        kg_dry = kg * dry_coeff
        print(f"{YELLOW}💡 Пересчёт: {kg} кг свежих → {kg_dry:.1f} кг сухих{RESET}\n")
    else:
        kg_dry = kg

    # Коэффициенты
    rates = {"hemp": 1.5, "bamboo": 1.6, "tree": 1.8}
    rate = rates.get(plant, 1.5)
    co2_saved = kg_dry * rate

    # Эквиваленты
    trees_equivalent = co2_saved / 22
    cars_equivalent = co2_saved / 4600

    # Номер сертификата (последовательный)
    cert_id = get_next_cert_number()
    date = datetime.now().strftime("%Y-%m-%d %H:%M")

    # ЦВЕТНОЙ ВЫВОД
    print(f"\n{GREEN}{BOLD}{'='*60}{RESET}")
    print(f"{GREEN}{BOLD}🏡 CARBON CERTIFICATE FARM{RESET}")
    print(f"{GREEN}{BOLD}✅ СЕРТИФИКАТ №{cert_id}{RESET}")
    print(f"{GREEN}{BOLD}{'='*60}{RESET}")
    print(f"{BLUE}🌱 Растение:{RESET} {plant.title()}")
    print(f"{BLUE}⚖️ Сухая биомасса:{RESET} {kg_dry:.1f} кг")
    print(f"{GREEN}💨 Запечатано CO₂:{RESET} {co2_saved:.1f} кг")
    print(f"{YELLOW}🌍 Эквивалент:{RESET} {co2_saved/1000:.3f} т CO₂")
    print(f"{BLUE}🌳 Деревьев (на 1 год):{RESET} {trees_equivalent:.1f} шт.")
    print(f"{RED}🚗 Автомобилей (на 1 год):{RESET} {cars_equivalent:.2f} шт.")
    print(f"{BLUE}📅 Дата:{RESET} {date}")
    print(f"{GREEN}{BOLD}{'='*60}{RESET}\n")

    # ГРАФИКИ
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    axes[0].bar(['Ваш вклад', '1 дерево'], [co2_saved, 22], color=['#2ecc71', '#95a5a6'])
    axes[0].set_ylabel('CO₂ (кг)')
    axes[0].set_title(f'🌳 Эквивалент: {trees_equivalent:.1f} деревьев', fontweight='bold')
    axes[0].grid(axis='y', alpha=0.3)

    plants = ['Конопля', 'Бамбук', 'Дерево']
    co2_values = [kg_dry * 1.5, kg_dry * 1.6, kg_dry * 1.8]
    colors = ['#27ae60', '#f39c12', '#8e44ad']
    axes[1].barh(plants, co2_values, color=colors)
    axes[1].set_xlabel('Запечатано CO₂ (кг)')
    axes[1].set_title('📊 Сравнение растений', fontweight='bold')
    axes[1].grid(axis='x', alpha=0.3)

    plt.tight_layout()
    plt.show()

    # СОХРАНЕНИЕ В ФАЙЛЫ
    # Текстовый файл
    txt_filename = f"carbon_cert_{cert_id}.txt"
    with open(txt_filename, 'w', encoding='utf-8') as f:
        f.write(f"🏡 CARBON CERTIFICATE FARM\n")
        f.write(f"{'='*60}\n")
        f.write(f"СЕРТИФИКАТ №{cert_id}\n")
        f.write(f"{'='*60}\n\n")
        f.write(f"Растение: {plant.title()}\n")
        f.write(f"Сухая биомасса: {kg_dry:.1f} кг\n")
        f.write(f"Запечатано CO₂: {co2_saved:.1f} кг\n")
        f.write(f"Эквивалент: {co2_saved/1000:.3f} т CO₂\n")
        f.write(f"Деревьев (на 1 год): {trees_equivalent:.1f} шт.\n")
        f.write(f"Автомобилей (на 1 год): {cars_equivalent:.2f} шт.\n")
        f.write(f"Дата: {date}\n\n")
        f.write(f"{'='*60}\n")
        f.write(f"Этот сертификат подтверждает вклад в борьбу с изменением климата.\n")
        f.write(f"Carbon Certificate Farm - https://github.com/yourusername/carbon-calc\n")

    # JSON файл
    json_filename = f"carbon_cert_{cert_id}.json"
    json_data = {
        "certificate_id": cert_id,
        "farm": "Carbon Certificate Farm",
        "plant": plant,
        "biomass_kg": kg_dry,
        "co2_saved_kg": co2_saved,
        "trees_equivalent": trees_equivalent,
        "cars_equivalent": cars_equivalent,
        "date": date
    }
    with open(json_filename, 'w', encoding='utf-8') as f:
        json.dump(json_data, f, ensure_ascii=False, indent=2)

    print(f"{GREEN}✅ Файлы созданы:{RESET}")
    print(f"   📄 {txt_filename}")
    print(f"   📄 {json_filename}")

    # Автоматическое скачивание
    download = input(f"\n{YELLOW}💾 Скачать файлы? (y/n): {RESET}").strip().lower()
    if download == 'y':
        print(f"{BLUE}📥 Скачивание...{RESET}")
        files.download(txt_filename)
        files.download(json_filename)
        print(f"{GREEN}✅ Готово!{RESET}")

    return {
        "cert_id": cert_id,
        "co2_saved": co2_saved,
        "trees": trees_equivalent
    }

# Запуск
carbon_certificate()
```
